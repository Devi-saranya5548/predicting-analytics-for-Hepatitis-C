```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/hcvdat0.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isna().sum()
```

```python
df.shape
```

```python
df.head(4)
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df.Sex.replace(('m','f'),(0,1),inplace=True)
```

```python
df.head(3)
```

```python
df['Category'].value_counts()
```

```python
df['Category']=[0 if i=="0=Blood Donor" or i=="0s=suspect Blood Donor" else 1
if i=="1=Hepatitis" else  2 if i=="2=Fibrosis" else 3 for i in df['Category']]
```

```python
df['Category'].value_counts()
```

```python
df.head(4)
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split

X=df.drop("Category",axis=1)
y=df['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

```python
#Linear Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```

```python
#Decision Trees
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

model = DecisionTreeClassifier() # [2]
model.fit(X_train, y_train) # [2]
y_pred = model.predict(X_test) # [2]

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```

```python
#SVM
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report

model = svm.SVC(kernel='linear') # Linear Kernel

model.fit(X_train, y_train) # [2]
y_pred = model.predict(X_test) # [2]

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```

```python
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

model = GaussianNB()
model.fit(X_train, y_train) # [1, 2]
y_pred = model.predict(X_test) # [1, 2]

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create a Random Forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42) # You can adjust n_estimators

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```
